# Background and Prior Work

*Fill in your background and prior work here (~500 words). You are welcome to use additional subheadings. You should also include a paragraph describing each dataset and how you'll be using them.* 

Motor neuron subtypes in Drosophila can innervate the same muscle but produce different functional outputs because they differ in synaptic organization and presynaptic release dynamics. At the Drosophila larval neuromuscular junction (NMJ), glutamatergic motor neurons are commonly grouped into tonic (type Ib) and phasic (type Is) classes. These two inputs can co-innervate the same muscle fibers, but they differ in properties such as effective strength, short-term plasticity (depression vs facilitation tendencies), and calcium-dependent release sensitivity. Phasic motoneurons are characterized by transient bursts of activity which create stronger depressing synapses with multiple muscle fibers. On the other hand, tonic motoneurons mediate sustained activity and form weaker facilitating synapses onto singular muscles (Grimby and Hannerz; Jetti et. al). Understanding how these functional differences arise is important because it connects cellular identity to circuit-level behavior and provides a tractable model for synaptic diversity.

A major question is whether tonic vs phasic identity is supported by distinct molecular programs. Presynaptic mechanisms like vesicle pool dynamics, active zone organization, and calcium coupling can strongly shape release probability and plasticity (Glasgow et. al). Prior work has shown physiological and nanoscale differences between tonic and phasic terminals, suggesting that gene expression differences could contribute to subtype-specific synaptic behavior. In other systems, changes in calcium channel composition and presynaptic regulatory proteins can shift synapses toward facilitation or depression, so transcriptomic differences are a plausible upstream driver of tonic/phasic functional diversity (Mochida).

To directly connect gene expression to neuron subtype, Patch-seq approaches combine electrophysiological recording with RNA sequencing from the same single neuron. Jetti et al. (2023) profiled Drosophila tonic and phasic motor neurons and reported subtype-associated molecular programs at isoform resolution. This makes the dataset well-suited for a data science question: can we detect tonic vs phasic structure using expression patterns alone, and can we identify candidate isoforms most responsible for that separation?

Detection of these differences leads to the question of what those structural variences are. One method of identifying structural differences in the synaptic terminal is by analyzing miniature postsynaptic currents (mPSCs) and using that information to infer about the presynaptic neuron. Miniature postsynaptic currents (mPSCs) are miniature responses evoked by the spontaneous fusion of presynaptic vesicles (Glasgow et. al). The measurement of these “minis” can uncover information about the quantal size and release probability (P{r}) in presynaptic neurons (Dervinis and Major). Differences in the amplitude of mPSCs can in the right conditions be looked at as a proxy to differences in quantal size, while variations in frequency are often linked to the quantity and density of low-threshold T-type calcium channels like Cav3.1 and Cav3.2 which mediate the stochastic calcium influx responsible for spontaneous vesicular fusion and subsequent mPSCs. These calcium channels are largely what determine the probability of release in the absence of an action potential. Of note, is the influence of the pool of releasable quanta on changes of frequency of mPSCs, however this influence is largely mitigated by the use of pharmacological blockers (Glasgow et. al).

Standard data collection procedures employ the use of Tetrodotoxin (TTX) to block voltage gated sodium channels in the presynaptic neuron, pharmacologically preventing the generation of action potentials in the presynaptic neuron ensuring that only spontaneous vesicular fusion events are recorded from the postsynaptic neuron (Glasgow et. al). The mPSCs are obtained via voltage clamp recordings of the postsynaptic cell. The dataset compiled by He et. al additionally employed the used of the neurotoxin BoNTC- to gather information on unsilenced Drosophila NMJs ("wild"), tonic Ib silenced with phasic Is normal NMJs, and phasic Is silenced with tonic Ib normal NMJs. This data provides the opportunity to compare phasic and tonic motoneurons with the same external and postsynaptic conditions, allowing for less noise when analyzing the potential presynaptic structural implications of the mPSC.

```python
#necessary packages
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%pip install seaborn
import seaborn as sns

from sklearn.decomposition import PCA
from scipy.stats import ttest_ind

#importing the data
#genes = pd.read_csv('1b1sm_intCt.txt', sep = '\t') 
psmRNA = pd.read_csv('1b1sm_isoform_intCt.txt', sep = '\t')
psmRNA.set_index('transcript_id', inplace=True)

#display the first 5 rows
psmRNA.head()
```

## Data Wrangling

Describe your data wrangling steps here.


1. **Loaded** the Patch-seq isoform count table into a pandas DataFrame (rows = transcript/isoform IDs, columns = single-cell samples).
2. **Log-transformed** expression values using log2(count + 1) to stabilize variance and make extreme counts easier to compare.
3. **Filtered** to the top 50 most variable isoforms (highest variance across cells) to focus visualizations on the features most likely to separate cell types.
4. **Visualized** the resulting subset using a heatmap and hierarchical clustering (clustermap) to see whether samples form expression-based groups.

```python
#convert the data to log2 to scale the data
psmRNA_log = np.log2(psmRNA+1)

#Filter for the top 50 most varible genes
top_genes = psmRNA_log.var(axis=1).sort_values(ascending=False).head(50).index
psmRNA_subset = psmRNA_log.loc[top_genes]

#create and display heatmap
plt.figure(figsize=(20,10))
sns.heatmap(psmRNA_subset,
            cmap='coolwarm',
            xticklabels=True,
            yticklabels=True)
plt.title('Gene Expressions Heat Map (Log2)')
plt.xlabel('Samples')
plt.ylabel('Transcript ID')
plt.xticks(rotation=30)
plt.show()

sns.clustermap(psmRNA_subset,
            cmap='coolwarm',
            standard_scale=0,
            figsize=(12,12))
plt.show()
    
```

```python
# ---- Label samples by subtype from column names ----
# ".1B" = tonic (Ib), ".1S" = phasic (Is)
def infer_cell_type(colname: str):
    if ".1B" in colname:
        return "Tonic (Ib)"
    if ".1S" in colname:
        return "Phasic (Is)"
    return "Other"

cell_type = pd.Series({c: infer_cell_type(c) for c in psmRNA_log.columns}, name="cell_type")
cell_type.value_counts()
```

```python
from sklearn.decomposition import PCA

# ---- PCA on cells ----
# Use top 1000 variable isoforms for PCA (more stable than only 50)
top_for_pca = psmRNA_log.var(axis=1).sort_values(ascending=False).head(1000).index
X = psmRNA_log.loc[top_for_pca].T  # rows=cells, cols=isoforms

pca = PCA(n_components=2, random_state=0)
pcs = pca.fit_transform(X)

pca_df = pd.DataFrame(pcs, columns=["PC1", "PC2"], index=X.index)

plt.figure(figsize=(7,5))
sns.scatterplot(data=pca_df, x="PC1", y="PC2")
plt.title("PCA of isoform expression (top 1000 variable isoforms)")
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
plt.tight_layout()
plt.show()
```

**Interpretation:** PCA compresses each cell’s isoform expression profile into two axes capturing the largest sources of variation. If cells form distinct clusters, it suggests consistent global transcriptomic differences between subgroups.

```python
pca_df["cell_type"] = cell_type.loc[pca_df.index].values

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="cell_type",
    palette={"Tonic (Ib)": "#E64B35", "Phasic (Is)": "#4DBBD5", "Other": "gray"},
    s=70,
    alpha=0.8
)
plt.axhline(0, ls="--", c="lightgray")
plt.axvline(0, ls="--", c="lightgray")
plt.title("PCA of Isoform Expression Colored by Motor Neuron Subtype", fontsize=14)
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
plt.legend(title="Subtype")
plt.tight_layout()
plt.show()
```

**Interpretation:** If tonic (Ib) and phasic (Is) points separate in PCA space, this supports subtype-associated transcriptomic programs. Overlap suggests either shared programs or technical/biological variability.

```python
from scipy.stats import ttest_ind

# ---- Subset to tonic and phasic only ----
keep_cols = cell_type[cell_type.isin(["Tonic (Ib)", "Phasic (Is)"])].index
expr = psmRNA_log[keep_cols]
labels = cell_type.loc[keep_cols]

tonic_cols = labels[labels == "Tonic (Ib)"].index
phasic_cols = labels[labels == "Phasic (Is)"].index

tonic = expr[tonic_cols]
phasic = expr[phasic_cols]

# ---- Welch t-test per isoform ----
mean_tonic = tonic.mean(axis=1)
mean_phasic = phasic.mean(axis=1)
log2fc = mean_tonic - mean_phasic  # tonic - phasic

pvals = []
for tx in expr.index:
    _, p = ttest_ind(tonic.loc[tx], phasic.loc[tx], equal_var=False, nan_policy="omit")
    pvals.append(p)

de = pd.DataFrame({
    "mean_tonic": mean_tonic,
    "mean_phasic": mean_phasic,
    "log2FC_tonic_minus_phasic": log2fc,
    "p_value": pvals
}).sort_values("p_value")

# ---- Benjamini–Hochberg FDR ----
m = len(de)
ranks = np.arange(1, m + 1)
de["fdr_bh"] = (de["p_value"] * m / ranks).clip(upper=1.0)

# quick summary
(de["fdr_bh"] <= 0.05).sum(), (de["fdr_bh"] <= 0.10).sum()
```

```python
# Prepare data for volcano plot
de_plot = de.copy()
de_plot["neglog10_p"] = -np.log10(de_plot["p_value"].replace(0, np.nextafter(0, 1)))
de_plot["significant"] = de_plot["fdr_bh"] <= 0.05

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=de_plot,
    x="log2FC_tonic_minus_phasic",
    y="neglog10_p",
    hue="significant",
    palette={True: "#E76F51", False: "#BDBDBD"},
    alpha=0.7,
    s=20,
    edgecolor=None
)

plt.axhline(-np.log10(0.05), linestyle="--", color="black", linewidth=1)
plt.xlabel("log2 fold change (tonic − phasic)")
plt.ylabel("−log10(p-value)")
plt.title("Volcano plot of isoform differential expression")
plt.legend(title="FDR ≤ 0.05")
plt.tight_layout()
plt.show()
```

**Interpretation:** Isoforms with large |log2FC| and low p-values are the strongest candidates differentiating tonic vs phasic neurons. Positive log2FC means higher expression in tonic; negative means higher in phasic. We also report FDR-adjusted significance because thousands of isoforms are tested.

```python
#pip install python-docx pandas

from docx import Document

electro_doc = Document('media-1 (1).docx')
mPSC_table = electro_doc.tables[0]
mPSC_td = []
for row in mPSC_table.rows:
    row_data = [cell.text for cell in row.cells]
    mPSC_td.append(row_data)

mPSC_data = pd.DataFrame(mPSC_td[1:], columns = mPSC_td[0])

mPSC_data.loc[mPSC_data['Label'] == 'Ib only (Is> BoNT)', 'Label'] = 'Ib only '
mPSC_data.loc[mPSC_data['Label'] == 'Is only (Ib>BoNT)', 'Label'] = 'Is only '

print(mPSC_data)

#for col in mPSC_data.columns:
#    mPSC_data[col] = pd.to_numeric(mPSC_data[col], errors='coerce')

cleaned_mPSC = mPSC_data.copy()

cleaned_mPSC['mEPSP_amp_mean'] = (cleaned_mPSC['mEPSP amplitude (mV)'].str.extract(r'([\d\.]+)').astype(float))
cleaned_mPSC['mEPSP_freq_mean'] = (cleaned_mPSC['mEPSP frequency (Hz)'].str.extract(r'([\d\.]+)').astype(float))

cleaned_mPSC['[Ca2+] (mM)'] = pd.to_numeric(cleaned_mPSC['[Ca2+] (mM)'])

amp_summary = (cleaned_mPSC.groupby(['[Ca2+] (mM)', 'Label'])['mEPSP_amp_mean'].mean().reset_index())
freq_summary = (cleaned_mPSC.groupby(['[Ca2+] (mM)', 'Label'])['mEPSP_freq_mean'].mean().reset_index())

fig, ax = plt.subplots()

for neuron in ['wild type', 'Ib only ', 'Is only ']:
    sub = amp_summary[amp_summary['Label'] == neuron]
    ax.plot(sub['[Ca2+] (mM)'], sub['mEPSP_amp_mean'], marker='o', linewidth=2, label=neuron)

ax.set_xlabel('Extracellular Ca²⁺ (mM)')
ax.set_ylabel('mEPSP amplitude (mV)')
ax.set_title('mEPSP amplitude across Ca²⁺ concentrations')
ax.legend()

plt.tight_layout()
plt.show()
```

### Overview of Analysis Pipeline

We performed a multi-step analysis to test whether tonic (Ib) and phasic (Is) motor neurons differ in isoform-level gene expression.

1. Isoform counts were log2-transformed to stabilize variance across cells.
2. Highly variable isoforms were identified to focus on biologically informative features.
3. Unsupervised clustering and heatmaps were used to assess global structure in the data.
4. PCA was applied to evaluate whether cells separate by subtype in low-dimensional space.
5. Differential isoform expression analysis was performed between tonic and phasic groups.
6. Multiple testing correction was applied to identify robust subtype-associated isoforms.

### Differential Expression Results

Differential expression testing identified multiple isoforms with statistically significant differences between tonic and phasic motor neurons after FDR correction.

Isoforms with large absolute log2 fold-changes and low adjusted p-values represent the strongest candidates for subtype-specific expression.  
Positive log2 fold-change values indicate higher expression in tonic neurons, while negative values indicate enrichment in phasic neurons.

These results suggest that subtype identity is associated with distinct isoform-level transcriptional programs.
