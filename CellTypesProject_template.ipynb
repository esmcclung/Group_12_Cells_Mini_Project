## Setup
*To get started, we need to import necessary packages. These will assist us in creating visuals and making the data easier to analyze. Next we imported our data into a dataframe and displayed the first 5 rows.*

```python
#necessary packages
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#importing the data
#genes = pd.read_csv('1b1sm_intCt.txt', sep = '\t') 
psmRNA = pd.read_csv('1b1sm_isoform_intCt.txt', sep = '\t')
psmRNA.set_index('transcript_id', inplace=True)
#display the first 5 rows
psmRNA.head()
```

## Data Wrangling

Describe your data wrangling steps here.

```python
#convert the data to log2 to scale the data
psmRNA_log = np.log2(psmRNA+1)

#Filter for the top 50 most varible genes
top_genes = psmRNA_log.var(axis=1).sort_values(ascending=False).head(50).index
psmRNA_subset = psmRNA_log.loc[top_genes]

#create and display heatmap
plt.figure(figsize=(15,10))
sns.heatmap(psmRNA_subset,
            cmap='coolwarm',
            xticklabels=True,
            yticklabels=True)
plt.show()
```
